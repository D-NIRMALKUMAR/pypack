```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Reshape
```

```python

# Define the dimensions of the image
image_width = 64
image_height = 64
```

```python
# Define the size of the random noise vector
latent_dim = 100
```

```python
# Define a simple neural network model
model = Sequential([
    Dense(128, input_dim=latent_dim, activation='relu'),
    Dense(image_width * image_height, activation='sigmoid'),
    Reshape((image_width, image_height))
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# Generate random noise as input
num_samples = 1
random_noise = np.random.normal(size=(num_samples, latent_dim))
```

```python
# Generate black and white images using the model
generated_images = model.predict(random_noise)
```

```python
# Visualize the generated black and white image
plt.imshow(generated_images[0], cmap='gray')
plt.axis('off')
plt.show()
```
