```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('M.csv')

# Preprocess the data
features = data[['X1', 'X2']]


# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features)
data['Clusters'] = kmeans.labels_

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X1', y='X2', data=data, hue='Clusters', palette='tab10')

plt.title('K-Means Clustering of M')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(title='Cluster')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# List to hold WCSS values
wcss = []

# Loop over a range of cluster numbers
for i in range(1, 8):  # Test 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(features)  # scaled_features is the standardized data
    wcss.append(kmeans.inertia_)
    
    
print(wcss)
# Plotting the elbow graph
plt.figure(figsize=(8, 5))
plt.plot(range(1, 8), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# List to hold silhouette scores
silhouette_scores = []

# Loop over a range of cluster numbers
for i in range(2, 8):  # Minimum number of clusters is 2
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(features)
    labels = kmeans.labels_
    score = silhouette_score(features, labels)
    silhouette_scores.append(score)
print(silhouette_scores)
# Plotting silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 8), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
